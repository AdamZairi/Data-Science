```python
# Pandas Data Science Library

import pandas as pd
import numpy as np
print (pd.__version__)
```

```python
# Pandas Series 
# Series is 1D Labelled Array

s = pd.Series([10,20,30,40,50],dtype='int32')
print(type(s))
print(s)

# numpy array only holds numbers
# pandas Series can hold numbers + labels(index)
```

```python
info = pd.Series(['Jia',21,'Tourism','Perak'],index=['name','age','course','city'])
print(info)
print(info['name'])
```

```python
info = pd.Series(['Jia',21,'Tourism','Perak'],index=['name','age','course','city'])
print(info)
print(info['name'])        # Indexing Method
print(info.get('age'))     # get() method
print(info.iloc[3])        # iloc -----> indexs location

# Series is somewhat a combination of dictionary + array.
```

```python
# values property
# prints only the values 

print(info.values)
print(info.values[2])
```

```python
# indexing Slicing and loops for Series

s = pd.Series([100,200,300,400,500,600],index=['v1','v2','v3','v4','v5','v6'])
print(s)
 
# 100  200   300   400    500     600
# 0     1     2     3      4       5
# -6    -5    -4    -3     -2      -1
# v1     v2   v3     v4    v5      v6
 
# Indexing in Series can be done by using [] or iloc[] or loc[]
print('\nIndexing In Series')
print(s.iloc[5])    # for index numbers
print(s.loc['v2'])  # loc used for labels
print(s['v3'])      # [] used for labels
print(s.iloc[-6])  

print('\nSlicing in Series')
print(s[0:3])
print(s['v1':'v5'])

# print 100, 400, 600
print(f'\n''Printing specific values')
print(s[['v1','v4','v6']])
print(s[['v1','v4','v6']].values)

# Conditional Selection
print(f'\n''Condition Selection')
print(s[s%4==0])
```

```python
# For loops in Pandas Series

for i in s:
    print(i)
print('\n')
for k,v in s.items():
    print(k,v)
print('\n')
```
