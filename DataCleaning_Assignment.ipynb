```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data_link = 'https://raw.githubusercontent.com/realpython/python-data-cleaning/refs/heads/master/Datasets/BL-Flickr-Images-Book.csv'
data = pd.read_csv(data_link)
data.head(10)
```

```python
# Dropping columns that wont be usefull

to_drop = ['Edition Statement',
           'Corporate Author',
           'Corporate Contributors',
           'Former owner',
           'Engraver',
           'Contributors',
           'Issuance type',
           'Shelfmarks']

data.drop(to_drop, inplace=True, axis=1)
```

```python
data.head(10)
```

```python
# checking whether data in the colum is unique to be set as index
data['Identifier'].is_unique
```

```python
# replace existing index
data = data.set_index('Identifier')
data.head()
```

```python
# accessing record with loc[], can refer to the value of the Identifier
data.loc[218]
```

```python
data.info()
```

```python
# column that is possible to change from object to numeric
# Date of Publication

# look for date that has more that 4 digit
data[data['Date of Publication'].str.len() > 4]['Date of Publication'].head(20)
```

```python
# Remove the extra dates in square brackets, wherever present: 1879 [1878]
# Convert date ranges to their “start date”, wherever present: 1860-63; 1839, 38-54
# Completely remove the dates we are not certain about and replace them with NumPy’s NaN: [1897?]
# Convert the string nan to NumPy’s NaN value

# extract the correct date with regex
extr = data['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
# convert the column to from object numerical
data['Date of Publication'] = pd.to_numeric(extr)
data['Date of Publication'].dtype
```

```python
data.loc[data['Date of Publication'].isnull(), 'Date of Publication']
```

```python
# Percentage of missing data
perc_missing = data['Date of Publication'].isnull().sum()/len(data)*100
print(perc_missing)
```

```python
# Replacing data names.

pub = data['Place of Publication']
london = pub.str.contains('London')
oxford = pub.str.contains('Oxford')
plymouth = pub.str.contains('Plymouth')

data['Place of Publication'] = np.where(london, 'London',
                                        np.where(oxford, 'Oxford',
                                                 np.where(plymouth, 'Plymouth',
                                                          pub.str.replace('-', ' '))))
```

```python
data['Place of Publication'].head(10)
```

```python
#Changing Column Type.

print(data['Place of Publication'].dtypes)
data['Place of Publication'] = data['Place of Publication'].astype('category')
print(data['Place of Publication'].dtypes)
```

```python
# cleaning entire dataset using applymap function

import requests

data = "https://raw.githubusercontent.com/realpython/python-data-cleaning/refs/heads/master/Datasets/university_towns.txt"

response = requests.get(data)
text_data = response.text

# Print first 20 lines
for line in text_data.splitlines()[:20]:
    print(line)
```

```python
# generating a dataframe from the txt file
# based on the txt file pattern, town name will have [edit]
# followed by university
unis_towns = []
for line in text_data.splitlines():
  if '[edit]' in line:
    state = line
  else:
    unis_towns.append((state,line))

unis_towns[:5]
```

```python
# converting the data structure to a dataframe

unis_towns_df = pd.DataFrame(unis_towns, columns=['State', 'Region Name'])
unis_towns_df
```

```python
unis_towns_df.info()
```

```python
def get_citystate(item):
  if ' (' in item:
    return item[:item.find(' (')]
  elif '[' in item:
    return item[:item.find('[')]
  else:
    return item
```

```python
# clean the data from symbol
unis_towns_df =  unis_towns_df.applymap(get_citystate)
unis_towns_df.head(20)
```

```python
# Renaming columns and Skipping rows
data = 'https://raw.githubusercontent.com/realpython/python-data-cleaning/refs/heads/master/Datasets/olympics.csv'
olympics_data = pd.read_csv(data)
olympics_data.head()

# skip first row to change header
# column 0 - country
# column 1,6,11 - type of games
# column 2,3,4 etc - gold siler bronze medal
```

```python
# Changing the header by skipping the first row.

olympics_data = pd.read_csv(data, header=1)
olympics_data.head()
```

```python
# rename column using rename() function
columnames = {'Unnamed: 0': 'Country',
              '? Summer': 'Summer Olympics',
              '01 !': 'Gold',
              '02 !': 'Silver',
              '03 !': 'Bronze',
              '? Winter': 'Winter Olympics',
              '01 !.1': 'Gold.1',
              '02 !.1': 'Silver.1',
              '03 !.1': 'Bronze.1',
              '? Games': '# Games',
              '01 !.2': '# Gold.2',
              '02 !.2': '# Silver.2',
              '03 !.2': '# Bronze.2'}

olympics_data = olympics_data.rename(columns=columnames)
olympics_data.head()
```

```python
olympics_data.info()
```

```python
# changing/cleaning country name
def get_country(item):
  if '[' in item:
    return item[:item.find('[')]
  else:
    return item

# clean the data from symbol
olympics_data['Country'] =  olympics_data['Country'].apply(get_country)
olympics_data.head(20)
```
