```python
import numpy as np
import pandas as pd
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [4,6,2,7,2,8]
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')
plt.show()
```

```python
from matplotlib import lines
# Adding styles to a graph

from matplotlib import style
style.available
style.use('ggplot')
x1 = [5,8,10]
y1 = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x1,y1,color='blue',linewidth=5,linestyle='solid',label='Product 1')
plt.plot(x2,y2,color='red',linewidth=5,linestyle='solid',label='Product 2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph with styles')
plt.legend()
plt.grid(True,color='lightblue')
plt.show()
```

```python
# bar chart
# bar()

plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Data 1')
plt.bar([2,4,6,8,10],[8,6,2,5,6],label='Data 2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
dept = ['HR','IT','ENG']
st_2024 = [12,24,34]
st_2025 = [18,28,40]
plt.bar(dept,st_2025,color='cyan',label='2025')
plt.bar(dept,st_2024,color='magenta',label='2024')
plt.xlabel('Department')
plt.ylabel('Staff')
plt.title('Company x')
plt.legend()
plt.show()
```

```python
# horizontal bar graph
# barh()

products = ['Comp','lappy','mouse','keuboard','printers']
quantity = [100,300,500,400,50]
plt.barh(products,quantity)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
# pie plot

cars = ['Toyota','Honda','Kia','Lambo','Perodua']
data = [20,40,15,5,75]
color = ['#244F8A','#9FAB37','#6A207D','#912020','#D1CDCD']
plt.pie(data,labels=cars,colors=color,
        startangle=260,
        explode=(0,0,0,0.1,0),
        autopct='%1.2f%%')
plt.title('Most used cars')
plt.show()
```

```python
# Generate a graph using a Dataset
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Grade_Set_1.csv')
data
```

```python
plt.scatter(data.Hours_Studied,data.Test_Grade,color='red')
plt.plot(data.Hours_Studied,data.Test_Grade,color='green',label='Data from dataset')
plt.xlabel('Hours_Studied',fontsize=12,fontweight='bold')
plt.ylabel('Test_Grade',fontsize=12,fontweight='bold')
plt.title('Hours_Studied VS Test_Grade',fontsize=12,fontweight='bold')
plt.legend()
plt.show()
```

```python
# xticks() yticks()

from numpy import random
x = random.randint(50,size=(10))
y = random.randint(50,size=(10))
plt.plot(x,y,color='#383896')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks(np.arange(1,55,5))
plt.yticks(np.arange(1,55,5))
plt.title('xticks & yticks')
plt.show()
```

```python
# subplots

# plot 1
x1 = np.random.randint(10,size=(4))
y1 = np.random.randint(6,size=(4))
plt.subplot(1,2,1)
plt.plot(x1,y1,color='red')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot 1')

# plot 2

x1 = np.random.randint(10,size=(6))
y1 = np.random.randint(6,size=(6))
plt.subplot(1,2,2)
plt.plot(x1,y1,color='yellow')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot 2')

plt.show()
```
