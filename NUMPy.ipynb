```python
# copy() vs view()
# copy()
# create the copy of an array in a different memory location

# if you make any changes in copied_array it won't affect the original array
# if you make any changes in original_array it won't affect the copied array

arr = np.array([1,2,3,4,5])
copy_arr = arr.copy()
arr[4] = 10
print(f'Original array :{arr}')
print('Copied Array :' ,copy_arr)
```

```python
# view()
# is the representation of the original array
# it shares the same memory location
# if you make any changes in copied_array it will affect the original array

arr = np.array([1,2,3,4,5])
copy_arr = arr.view()
arr[4] = 10

print(f'Original array :{arr}')
print(f'Copied Array : {copy_arr}')

# shares_memory() --> returns TRUE if arrays are sharing same memory location

print(f'\n',np.shares_memory(arr,copy_arr))
```

```python
# not divisibe by 2
arr = np.array([1,4,2,7,9,3,5,8,12,14,13,9,6,5,11,7])
arr%2!=0
arr = arr[arr%2!=0] 
print(arr)
```

```python
# not divisibe by 2 and should be greater than 7
arr = np.array([1,4,2,7,9,3,5,8,12,14,13,9,6,5,11,7])
arr%2!=0
arr>7
arr = arr[(arr%2!=0)&(arr>7)]
print(arr)

```

```python
# np.where(expression)

arr = np.array([1,4,2,7,9,3,5,8,12,14,13,9,6,5,11,7])
res = np.where((arr>=7) & (arr%2!=0))
print(arr[res])
print(f"Index Number :",res)
```

```python
arr = np.array([10, 32, 30, 50, 20, 82, 91, 45])

# Find indices where value is 30
indices = np.where(arr == 30)
print(f"Indices where value is 30: {indices}")

# Find indices where values are even
even_indices = np.where(arr % 2 == 0)
print(f"Indices of even values: {even_indices}")

# Conditional selection
new_arr = np.where(arr > 40 ,arr * 2 ,arr)
print(f"Array after conditional selection: {new_arr}")
```

```python
# Numpy Broadcasting
 
arr1 = np.array([5,8,10,12])
arr2 = np.array([3,7,6,5])
 
print(arr1+arr2)
print(arr1*arr2)
```

```python
# Broadcasting cannot perform arithmetic operations with different shapes


arr1 = np.array([5,8,10,12])
arr2 = np.array([3,7])
 
print(arr1+arr2)
print(arr1*arr2)
```

```python
# General Rules To Avoid Broadcasting
 
# a.) both array dimensions should be equal
# b.) if any of them is 1
 
arr1 = np.array([5,8,10,12])
arr2 = np.array([3])
 
print(arr1+arr2)
```
